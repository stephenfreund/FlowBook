# DataFrame Structural Attributes

This notebook demonstrates all the DataFrame attributes that are tracked
as structural reads. These are attributes that reveal information about
the DataFrame's structure (columns, rows, types).

## Column-Revealing Attributes
Access to these attributes records that you read the column structure:
- `df.columns`
- `df.keys()`
- `for col in df:` (iteration)
- `df.dtypes`
- `df.T`
- `df.axes`
- `df.values`
- `df.describe()`
- `df.to_dict()`
- `df.info()`
- `df.head()` / `df.tail()` / `df.sample()`
- `df.select_dtypes()`
- `df.to_records()`
- `df.memory_usage()`

## Row-Revealing Attributes
Access to these attributes records that you read the row structure:
- `df.index`
- `len(df)`
- `df.shape`
- `df.size`
- `df.empty`

```python
import pandas as pd
import numpy as np

%structural_tracking warn
```

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'salary': [50000.0, 60000.0, 70000.0]
})
```

## Column-Revealing Attributes

```python
# .columns - most direct way to read column structure
print("Columns:", list(df.columns))
```

```python
# .keys() - equivalent to .columns
print("Keys:", list(df.keys()))
```

```python
# Iterating over DataFrame iterates over columns
print("Iterating:")
for col in df:
    print(f"  {col}")
```

```python
# .dtypes - reveals columns and their types
print("Dtypes:")
print(df.dtypes)
```

```python
# .T (transpose) - reveals column structure
print("Transposed:")
print(df.T)
```

```python
# .describe() - summarizes all numeric columns
print("Description:")
print(df.describe())
```

```python
# .to_dict() - converts to dict, revealing columns as keys
print("As dict:")
print(df.to_dict())
```

## Row-Revealing Attributes

```python
# .index - reveals row structure
print("Index:", list(df.index))
```

```python
# len() - number of rows
print("Length:", len(df))
```

```python
# .shape - reveals both row and column counts
print("Shape:", df.shape)
```

```python
# .size - total number of elements
print("Size:", df.size)
```

```python
# .empty - whether DataFrame is empty
print("Empty:", df.empty)
```
