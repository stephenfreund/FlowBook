# Series Structural Attributes

This notebook demonstrates the Series attributes that are tracked
as structural reads. Series have simpler structure than DataFrames
but still have trackable attributes.

## Tracked Attributes
- `s.index` - the index (row labels)
- `len(s)` - number of elements
- `s.shape` - shape tuple
- `s.size` - number of elements
- `s.empty` - whether Series is empty
- `s.dtype` - data type
- `s.name` - Series name
- `s.values` - underlying values
- `s.describe()` - statistical summary
- `s.to_dict()` / `s.to_list()` - conversions

```python
import pandas as pd
import numpy as np

%structural_tracking warn
```

```python
# Create a Series with a custom index
s = pd.Series(
    [100, 200, 300, 400],
    index=['a', 'b', 'c', 'd'],
    name='values'
)
s
```

## Index and Length

```python
# .index - the row labels
print("Index:", list(s.index))
```

```python
# len() - number of elements
print("Length:", len(s))
```

```python
# .shape - shape tuple
print("Shape:", s.shape)
```

```python
# .size - number of elements
print("Size:", s.size)
```

```python
# .empty - is the Series empty?
print("Empty:", s.empty)
```

## Type and Name

```python
# .dtype - data type
print("Dtype:", s.dtype)
```

```python
# .name - Series name
print("Name:", s.name)
```

## Values and Conversions

```python
# .values - underlying numpy array
print("Values:", s.values)
```

```python
# .to_list() - convert to Python list
print("As list:", s.to_list())
```

```python
# .to_dict() - convert to dict
print("As dict:", s.to_dict())
```

```python
# .describe() - statistical summary
print("Description:")
print(s.describe())
```

## Iteration

```python
# Iterating over Series iterates over values
print("Iterating over values:")
for val in s:
    print(f"  {val}")
```
