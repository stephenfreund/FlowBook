```python
import numpy as np
import pandas as pd
```

```python
# Create a large DataFrame (1M rows)
np.random.seed(42)
n_rows = 1_000_000

df = pd.DataFrame({
    'id': np.arange(n_rows),
    'value_a': np.random.randn(n_rows),
    'value_b': np.random.randn(n_rows),
    'value_c': np.random.randn(n_rows),
    'category': np.random.choice(['A', 'B', 'C', 'D'], n_rows),
})
print(f'DataFrame shape: {df.shape}')
print(f'Memory usage: {df.memory_usage(deep=True).sum() / 1e6:.1f} MB')
```

```python
# Create derived columns
df['sum_ab'] = df['value_a'] + df['value_b']
df['product_bc'] = df['value_b'] * df['value_c']
df['ratio'] = df['value_a'] / (df['value_b'].abs() + 1e-10)
print(f'Added 3 columns, new shape: {df.shape}')
```

```python
# Create a second large array
large_array = np.random.randn(2000, 2000)
print(f'Array shape: {large_array.shape}')
print(f'Array memory: {large_array.nbytes / 1e6:.1f} MB')
```

```python
# Compute statistics by category
stats = df.groupby('category').agg({
    'value_a': ['mean', 'std', 'min', 'max'],
    'value_b': ['mean', 'std', 'min', 'max'],
    'sum_ab': ['mean', 'std'],
})
print(stats)
```

```python
# Matrix operations on large array
result_matrix = large_array @ large_array.T
eigenvalues = np.linalg.eigvalsh(result_matrix[:100, :100])
print(f'Result matrix shape: {result_matrix.shape}')
print(f'Top 5 eigenvalues: {eigenvalues[-5:]}')
```

```python
# Filter and transform
df_filtered = df[df['value_a'] > 0].copy()
df_filtered['normalized'] = (df_filtered['value_a'] - df_filtered['value_a'].mean()) / df_filtered['value_a'].std()
print(f'Filtered shape: {df_filtered.shape}')
```

```python
# Create another large structure - dictionary of arrays
data_dict = {
    f'col_{i}': np.random.randn(500_000) for i in range(20)
}
print(f'Dictionary with {len(data_dict)} arrays')
print(f'Total size: {sum(arr.nbytes for arr in data_dict.values()) / 1e6:.1f} MB')
```

```python
# Aggregate results
final_summary = {
    'df_rows': len(df),
    'df_filtered_rows': len(df_filtered),
    'array_size': large_array.shape,
    'dict_keys': len(data_dict),
    'stats_shape': stats.shape,
}
print('Final summary:')
for k, v in final_summary.items():
    print(f'  {k}: {v}')
```
