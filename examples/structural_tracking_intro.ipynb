# Structural Tracking Introduction

This notebook introduces the concept of structural tracking in DataFerret.

**Problem**: When you access DataFrame attributes like `.columns`, `.shape`, or `len(df)`,
you're reading information about the DataFrame's *structure*. If another cell later
changes that structure (adds columns, rows), your earlier cell's output becomes stale.

**Solution**: Structural tracking detects when you access structure-revealing attributes
and can warn or block changes that would affect those reads.

```python
import pandas as pd
import numpy as np
```

```python
# Create a simple DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35]
})
df
```

```python
# This cell reads the column structure
print(f"Columns: {list(df.columns)}")
print(f"Number of columns: {len(df.columns)}")
```

```python
# This cell reads the shape (both rows and columns)
print(f"Shape: {df.shape}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# This cell reads the row count
print(f"Number of rows: {len(df)}")
```

## The Problem

If we now add a column, the outputs of the cells above become incorrect.
They said we have 2 columns, but now we have 3!

```python
# Add a new column - this changes the structure!
df['salary'] = [50000, 60000, 70000]
df
```

## The Solution: Structural Tracking

With structural tracking enabled:
- **WARN mode**: You get warnings when structure changes affect earlier reads
- **ENFORCE mode**: Changes that affect earlier structural reads are blocked

Use the `%structural_tracking` magic command to control the mode.

```python
# Show current structural tracking mode
%structural_tracking
```
