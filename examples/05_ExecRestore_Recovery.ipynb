# EXEC-RESTORE: Recovering from Forward Contamination

This notebook demonstrates using **"Run with Previous Cell's state"** to
recover a contaminated cell without restarting the kernel.

## How to run

### Phase 1: Create the contamination

1. Run **Cell A** (`x = 10, y = 20`)
2. Run **Cell C** (`x = -1`) — skipping Cell B
3. Run **Cell B** (`total = x + y`) — reads `x`, which was overwritten by C

Cell B is marked **stale** (forward contamination). It computed `total = 19`
using C's value of `x`, but in notebook order it should have used A's value.

### Phase 2: Recover with EXEC-RESTORE

4. Right-click **Cell B** and select **"Run with Previous Cell's state"**.

## What to expect

- Cell B re-executes using Cell A's checkpoint (where `x = 10`), producing
  `total = 30`.
- Cell B becomes **fresh** again.
- Cell C becomes **stale** because B's execution changed variables that
  may affect C's assumptions.

## Why this is useful

Without EXEC-RESTORE, you would need to restart the kernel and re-run
everything from scratch. EXEC-RESTORE lets you fix one cell's state
without losing the rest of your work.
