# Comprehensive Structural Tracking Demo

This notebook provides a complete walkthrough of structural tracking features.

```python
import pandas as pd
import numpy as np
```

## Part 1: Magic Command Basics

```python
# Show current mode
%structural_tracking
```

```python
# Set to OFF - no tracking
%structural_tracking off
```

```python
# Set to WARN - track and warn
%structural_tracking warn
```

```python
# Set to ENFORCE - track and block violations
%structural_tracking enforce
```

```python
# Case insensitive
%structural_tracking WARN
```

## Part 2: DataFrame Structural Reads

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'department': ['Engineering', 'Sales', 'Engineering', 'Marketing'],
    'salary': [75000, 65000, 80000, 70000],
    'years': [3, 5, 2, 4]
})
df
```

```python
# Column-revealing attributes
print("=== Column Structure ===")
print(f"columns: {list(df.columns)}")
print(f"dtypes:\n{df.dtypes}")
print(f"keys: {list(df.keys())}")
```

```python
# Row-revealing attributes
print("=== Row Structure ===")
print(f"index: {list(df.index)}")
print(f"len: {len(df)}")
print(f"shape: {df.shape}")
print(f"size: {df.size}")
print(f"empty: {df.empty}")
```

```python
# Methods that reveal structure
print("=== Methods ===")
print(f"describe:\n{df.describe()}")
```

```python
# Iteration reveals columns
print("=== Iteration ===")
for col in df:
    print(f"Column: {col}")
```

## Part 3: Series Structural Reads

```python
s = pd.Series([10, 20, 30, 40, 50], 
              index=['a', 'b', 'c', 'd', 'e'],
              name='values')
s
```

```python
# Series structural attributes
print("=== Series Structure ===")
print(f"index: {list(s.index)}")
print(f"len: {len(s)}")
print(f"shape: {s.shape}")
print(f"dtype: {s.dtype}")
print(f"name: {s.name}")
```

## Part 4: Nested DataFrames

```python
# DataFrames in a dictionary
data = {
    'train': pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]}),
    'test': pd.DataFrame({'x': [7, 8], 'y': [9, 10]})
}

print(f"Train columns: {list(data['train'].columns)}")
print(f"Test shape: {data['test'].shape}")
```

```python
# DataFrames in a list
batches = [
    pd.DataFrame({'val': [1, 2]}),
    pd.DataFrame({'val': [3, 4]}),
    pd.DataFrame({'val': [5, 6]})
]

for i, batch in enumerate(batches):
    print(f"Batch {i} length: {len(batch)}")
```

## Part 5: Real-World Pattern - Feature Engineering

```python
# Set to ENFORCE for demonstration
%structural_tracking warn
```

```python
# Simulate loading data
raw_data = pd.DataFrame({
    'user_id': [1, 2, 3, 4, 5],
    'purchase_amount': [100, 250, 75, 300, 150],
    'visit_count': [5, 12, 3, 8, 6]
})
```

```python
# Check schema - this is a structural read!
print(f"Data schema: {list(raw_data.columns)}")
print(f"Data types:\n{raw_data.dtypes}")
```

```python
# Add computed features - OK in a new cell
raw_data['avg_purchase'] = raw_data['purchase_amount'] / raw_data['visit_count']
raw_data['high_value'] = raw_data['purchase_amount'] > 200
raw_data
```

```python
# Check final schema
print(f"Final columns: {list(raw_data.columns)}")
```

```python
# Reset to warn
%structural_tracking warn
print("Demo complete!")
```
