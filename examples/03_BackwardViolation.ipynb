# Backward Violation (Rejected Execution)

This notebook shows FlowBook **rejecting** a cell execution that would
silently corrupt an earlier cell's result.

## How to run

1. Run **Cell A** (`x = 10`)
2. Run **Cell B** (`y = x * 2`) — reads `x`, produces `y = 20`
3. Run **Cell C** (`x = 999`) — overwrites `x`

## What to expect

- Cells A and B execute normally and stay **fresh**.
- When you run Cell C, FlowBook detects a **backward violation**: C writes
  to `x`, which was already read by the earlier fresh Cell B. If C were
  allowed, B's result (`y = 20`) would be inconsistent with the new value
  of `x`.
- Cell C is **rejected and rolled back**. The namespace is restored to its
  state before C ran. An error message explains why.
- `x` remains `10` — the rollback undid `x = 999`.

## Why this matters

In a standard Jupyter notebook, running Cell C would silently change `x`
to `999`, making Cell B's output (`y = 20`) inconsistent. FlowBook prevents
this by catching the conflict and rolling back.

```python
# Cell A
x = 10
```

```python
# Cell B
y = x * 2
print(f'y = {y}')  # Should print 20
```

```python
# Cell C — this will be rejected!
x = 999
print(f'x = {x}')  # You will NOT see this output
```

```python
# Cell D — verify rollback
print(f'x = {x}')  # Still 10, not 999
```
