# File Tracking - CSV with Pandas

This notebook demonstrates FlowBook's file tracking with pandas CSV operations.
Both `pd.read_csv()` and `df.to_csv()` are tracked.

```python
import pandas as pd
import numpy as np
```

## Create and save a DataFrame to CSV

The file write will appear in the `writes` list as `file:sales_data.csv`.

```python
# Create sample sales data
df = pd.DataFrame({
    'product': ['Widget A', 'Widget B', 'Widget C', 'Widget D'],
    'price': [10.99, 24.99, 5.99, 49.99],
    'quantity': [100, 50, 200, 25]
})

# Save to CSV
df.to_csv('sales_data.csv', index=False)
print('Saved sales_data.csv')
df
```

## Read the CSV file

The file read will appear in the `reads` list as `file:sales_data.csv`.

```python
# Read the CSV back
loaded_df = pd.read_csv('sales_data.csv')
loaded_df
```

## Process and analyze the data

```python
# Calculate total revenue
loaded_df['revenue'] = loaded_df['price'] * loaded_df['quantity']
total_revenue = loaded_df['revenue'].sum()
print(f'Total revenue: ${total_revenue:.2f}')
loaded_df
```

## Update the source file

If we modify the CSV file, cells that read it become stale.

```python
# Add a new product and save
new_product = pd.DataFrame({
    'product': ['Widget E'],
    'price': [99.99],
    'quantity': [10]
})
updated_df = pd.concat([df, new_product], ignore_index=True)
updated_df.to_csv('sales_data.csv', index=False)
print('Updated sales_data.csv with new product')
```

## Save processed results to a different file

```python
# Save the processed data with revenue to a new file
loaded_df.to_csv('sales_with_revenue.csv', index=False)
print('Saved sales_with_revenue.csv')
```

## Cleanup

```python
import os
for f in ['sales_data.csv', 'sales_with_revenue.csv']:
    if os.path.exists(f):
        os.remove(f)
        print(f'Cleaned up {f}')
```
