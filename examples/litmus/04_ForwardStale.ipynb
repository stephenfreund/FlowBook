# Litmus Test: ForwardStale Predicate

**Formal definition:** `ForwardStale(R, W, W', i, j) ≝ j > i ∧ W'ᵢ ∩ Rⱼ ≠ ∅`

When cell i (re-)executes and writes to a location, any later cell j that reads that location becomes stale.

**Test scenario:**
1. Run A (writes x=1)
2. Run B (reads x)
3. Re-run A with x=100

**Expected:** After step 3, B is marked stale with reason `forward_stale` on location `x`, pointing to cell A.

```python
# Cell A: Writes x
# After running A→B, change this to x=100 and re-run A
x = 1
```

```python
# Cell B: Reads x
y = x + 1
print(f"y = {y}")
```
