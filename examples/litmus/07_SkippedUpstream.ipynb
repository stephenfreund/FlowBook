# Litmus Test: SkippedUpstream Staleness

**Behavior:** When you re-run an early cell, skipping an intermediate writer, cells that were reading from the intermediate become stale.

This ensures readers are notified that their data source changed.

**Test scenario:**
1. Run F (writes x=1)
2. Run G (reads x, writes x=2)
3. Run H (reads x=2 from G)
4. Re-run F (x goes back to 1)

**Expected:** H is marked stale with reason `skipped_upstream` because H was reading from G, but now x comes from F (G was skipped).

```python
# Cell F: Initial writer of x
x = 1
```

```python
# Cell G: Reads x, writes x (intermediate)
x = x + 1  # x becomes 2
```

```python
# Cell H: Reads x
y = x * 5
print(f"y = {y}  (10 if from G, 5 if from F)")
```
