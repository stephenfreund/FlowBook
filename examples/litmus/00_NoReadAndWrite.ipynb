# Litmus Test: NoReadAndWrite Predicate

**Formal definition:** `NoReadAndWrite(R, W, i) ≝ Rᵢ ∩ Wᵢ = ∅`

A cell cannot read AND write the same variable in a way that would create a conflict. This is handled implicitly by separating reads-before-writes from the write set.

**Note:** This predicate is satisfied implicitly because FlowBook tracks `reads_before_writes` - reads that happen before any write to that variable. Self-modifications like `x = x + 1` are valid.

**Test scenario:**
1. Run setup (writes x=1)
2. Run A (reads x, then writes x = x + 1)

**Expected:** A executes successfully. The read of x happens before A writes to x, so there's no conflict. FlowBook correctly distinguishes the read dependency from the write output.

```python
# Setup: Initialize x
x = 1
print(f"x = {x}")
```

```python
# Cell A: Reads x, then writes x (self-modification)
# This is allowed because the read happens BEFORE the write
x = x + 1
print(f"x = {x}  (should be 2)")
```

```python
# Verify: Re-run setup, then A should be marked stale
# because A reads from setup's x
print(f"Current x = {x}")
```
