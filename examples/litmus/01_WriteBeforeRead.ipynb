# Litmus Test: WriteBeforeRead Predicate

**Formal definition:** `WriteBeforeRead(R, W, i) ≝ Rᵢ ⊆ W_{1..i-1}`

Every variable that cell i reads must have been written by some earlier cell. This ensures no undefined variable access.

**Note:** This predicate is not strictly enforced by FlowBook - Python itself raises `NameError` for undefined variables. FlowBook focuses on detecting reproducibility violations.

**Test scenario:**
1. Run A (tries to read undefined variable `undefined_var`)

**Expected:** Python raises `NameError: name 'undefined_var' is not defined`. FlowBook does not need to enforce this separately.

```python
# Cell A: Tries to read undefined variable
# This will raise NameError (Python enforces WriteBeforeRead)
y = undefined_var + 1
```

```python
# Cell B: Correct usage - define before use
z = 10
w = z + 1
print(f"w = {w}")
```
