# Litmus Test: NoWriteAfterRead Predicate

**Formal definition:** `NoWriteAfterRead(R, W, i) ≝ Wᵢ ∩ R_{1..i-1} = ∅` (for clean cells)

Cell i cannot write to a location that an earlier CLEAN cell has already read. This would silently invalidate the earlier cell's result.

**Test scenario:**
1. Run setup (creates x=1)
2. Run A (reads x)
3. Run B (tries to write x)

**Expected:** B fails `NoWriteAfterRead` because B writes x which A (an earlier clean cell) read. B is REJECTED and the namespace is rolled back.

```python
# Setup: Initialize x
x = 1
```

```python
# Cell A: Reads x (run FIRST after setup)
y = x + 1
print(f"y = {y}")
```

```python
# Cell B: Writes x (WILL BE REJECTED)
x = 999
print(f"x = {x}")  # Won't see this
```

```python
# Verify: x should still be 1 (rollback)
print(f"x = {x}")
```
