# Litmus Test: WriteOverlap Staleness

**Behavior:** When two cells write the same location, re-running one marks the other stale.

This prevents inconsistent state where the "wrong" writer's value is in the namespace.

**Test scenario:**
1. Run A (writes x=1)
2. Run B (writes x=2)
3. Re-run A (writes x=1 again)

**Expected:** After step 3, B is marked stale with reason `write_overlap` on location `x`.

```python
# Cell A: Writes x
x = 1
print(f"A: x = {x}")
```

```python
# Cell B: Also writes x
x = 2
print(f"B: x = {x}")
```

```python
# Cell C: Check current value
print(f"Current x = {x}")
```
