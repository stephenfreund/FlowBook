# DataFrame Column Independence

Run A → B → C in order.

**Expected:** C is accepted — it modifies `score` while B only read `age`. Different columns, no conflict.

```python
import pandas as pd
df = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'age': [30, 25],
    'score': [85, 92]
})
```

```python
avg_age = df['age'].mean()
print(f"Average age: {avg_age}")
```

```python
df['adjusted_age'] = df['age'] * 1.1  # OK: different column
df
```
