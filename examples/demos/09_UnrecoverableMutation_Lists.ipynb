# Unrecoverable Mutations: Lists and Variables

This notebook illustrates the difference between **unrecoverable mutations** (in-place modification) and **recoverable writes** (rebinding) for lists and plain variables.

## What does "unrecoverable" mean?

An **unrecoverable mutation** is an in-place modification to a variable's contents (e.g., `x[i] = val`) that FlowBook cannot undo by simply re-running the cell. If a later cell also mutates the same object in place, deleting that later cell and re-running the earlier one will not restore the object to the expected state — because in-place mutations accumulate and there is no prior version to fall back to.

A **recoverable write** is a full rebinding (e.g., `x = [...]`) that replaces the entire object. FlowBook can always recover from these by re-running the cell, since re-execution recreates the value from scratch.

```python
x = [1, 2, 3, 4, 5, 6]
```

```python
x[4] = 99  # In-place mutation — UNRECOVERABLE
```

```python
x[5] = 88  # In-place mutation — UNRECOVERABLE
```

```python
print(x)
```

## The problem

If Cell D (`x[5] = 88`) is deleted and Cell C (`x[4] = 99`) is re-run, the value `x[5]` will still be `88` — it was mutated in place and there is no way to restore it to `6`. The mutation from the deleted cell persists in the live object.

This is because both cells modify the *same list object* in memory. Deleting a cell does not undo its side effects on shared mutable state.

## Recoverable Version

Using full rebinding instead of in-place mutation makes the write recoverable.

```python
x = [1, 2, 3, 4, 99, 88]  # Rebinding — RECOVERABLE
```

Rebinding is safe because re-running this cell always produces the same value for `x`, regardless of what other cells may have done. FlowBook can track and recover from rebindings by simply re-executing the cell.
