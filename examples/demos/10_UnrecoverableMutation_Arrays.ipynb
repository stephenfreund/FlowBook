# Unrecoverable Mutations: NumPy Arrays

This notebook demonstrates unrecoverable in-place mutations on NumPy arrays versus recoverable rebinding.

```python
import numpy as np
arr = np.zeros(10)
```

```python
#total = arr.sum()  # Reads arr
```

```python
arr[0] = 42  # In-place mutation — UNRECOVERABLE
```

```python
arr[5] = 7  # In-place mutation — UNRECOVERABLE
```

## The problem

Both Cell C and Cell D mutate `arr` in place. If Cell D is deleted and Cell C is re-run, `arr[5]` will still be `7` — the mutation from the deleted cell persists in the underlying array buffer.

NumPy arrays are mutable buffers. Element assignment (`arr[i] = val`) modifies the buffer directly without creating a new array object. FlowBook cannot undo these mutations by re-running the cell that performed them.

## Recoverable Version

Rebinding `arr` to a new array is always recoverable.

```python
arr = np.array([42, 0, 0, 0, 0, 7, 0, 0, 0, 0])  # Rebinding — RECOVERABLE
```

Re-running this cell always produces the exact same array, regardless of prior in-place mutations. FlowBook can safely track and recover from full rebindings.
