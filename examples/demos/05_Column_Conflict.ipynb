# DataFrame Column Conflict

Run A → B → C in order.

**Expected:** C is rejected — it reads and writes `age`, same column B read.

```python
import pandas as pd
df = pd.DataFrame({
    'name': ['Alice', 'Bob'],
    'age': [30, 25]
})
```

```python
df['age'] = 4
```

```python
avg_age = df['age'].mean()
print(f"Average age: {avg_age}")
```

```python
# REJECTED: reads age (for +1) then writes age, but B already read age
df['age'] = df['age'] + 1
```
