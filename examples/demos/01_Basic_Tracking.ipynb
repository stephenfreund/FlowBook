# Basic Dependency Tracking

Run A → B → C. Then change A and re-run it.

**Expected:** B and C turn stale (amber) because they depend on A's output.

```python
price = 100
```

```python
tax = price * 0.08
print(f"Tax: ${tax:.2f}")
```

```python
total = price + tax
print(f"Total: ${total:.2f}")
```
