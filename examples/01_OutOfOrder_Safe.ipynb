# Out-of-Order Execution: Safe

This notebook shows that running cells out of order is **fine** when there are no
conflicting variable dependencies.

## How to run

1. Run **Cell A** (imports)
2. Run **Cell C** (`z = 10`) — skipping Cell B
3. Run **Cell B** (`x = 5`)
4. Run **Cell D** (`result = x + z`)

## What to expect

All cells stay **fresh** (green). Even though you ran C before B, there is no
conflict because B and C write to independent variables (`x` and `z`). Cell D
reads both, but it runs last, so it sees correct values.

```python
# Cell A: setup
import math
```

```python
# Cell B
x = 5
```

```python
# Cell C
z = 10
```

```python
# Cell D
result = x + z
print(f'result = {result}')  # Should print 15
```
