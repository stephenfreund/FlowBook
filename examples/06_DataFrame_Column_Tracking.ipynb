# DataFrame Column-Level Tracking

FlowBook tracks reads and writes at the **column level** for DataFrames.
This means modifying one column won't conflict with cells that only use
other columns.

## How to run

### Phase 1: Safe column independence

1. Run **Cell A** — creates a DataFrame with columns `name`, `age`, `score`
2. Run **Cell B** — reads `df['age']`, computes mean age
3. Run **Cell C** — modifies `df['score']` (a different column)

Cell C is **accepted** because it only writes to `score`, and Cell B only
read `age`. No conflict.

### Phase 2: Column conflict

4. Run **Cell D** — modifies `df['age']` (the column Cell B read)

## What to expect

- Phase 1: All cells stay **fresh**. FlowBook sees that `score` and `age`
  are independent columns.
- Phase 2: Cell D is **rejected** because it writes to `df['age']`, which
  was read by the earlier fresh Cell B. This is a backward violation at
  the column level.

```python
# Cell A
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Carol'],
    'age': [30, 25, 35],
    'score': [85, 92, 78]
})
df
```

```python
# Cell B — reads only the 'age' column
mean_age = df['age'].mean()
print(f'Mean age: {mean_age}')
```

```python
# Cell C — writes only 'score' (independent of 'age')
df['score'] = df['score'] * 1.1
print('Scores adjusted')
df
```

```python
# Cell D — writes 'age' (conflicts with Cell B!)
df['age'] = df['age'] + 1
print('Ages incremented')
df
```
