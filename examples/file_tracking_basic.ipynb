# File Tracking - Basic Examples

This notebook demonstrates FlowBook's file tracking capabilities.
File reads and writes are tracked and shown in the metadata with a `file:` prefix.

## Writing a file

When you write to a file, it appears in the `writes` list as `file:filename`.

```python
# Write some data to a file
with open('example_data.txt', 'w') as f:
    f.write('Hello, FlowBook!\n')
    f.write('Line 2\n')
    f.write('Line 3\n')
```

## Reading a file

When you read from a file that was written in a previous cell,
it appears in the `reads` list as `file:filename`.

Note: Only "read-before-write" files are tracked in reads -
files that are read and then written in the same cell are not counted as reads.

```python
# Read the file we created
with open('example_data.txt', 'r') as f:
    content = f.read()
print(content)
```

## File staleness

If we modify the file, cells that read it become stale,
just like variable dependencies.

```python
# Modify the file - this should make the reading cell stale
with open('example_data.txt', 'w') as f:
    f.write('Modified content!\n')
```

## Cleanup

```python
import os
if os.path.exists('example_data.txt'):
    os.remove('example_data.txt')
    print('Cleaned up example_data.txt')
```
