# FlowBook: Getting Started

This notebook demonstrates the basics of FlowBook's reproducibility tracking.

**Key concepts:**
- FlowBook tracks which variables each cell reads and writes
- It detects when cells become "stale" (need re-execution)
- It prevents backward mutations that break reproducibility

**Instructions:** Run the cells in order A -> B -> C, then try running them out of order.

## Cell A: Define a variable

```python
x = 10
```

## Cell B: Use the variable

```python
y = x * 2
print(f"y = {y}")
```

## Cell C: Use the result

```python
z = y + 5
print(f"z = {z}")
```

---

## Try it yourself!

1. **Run A -> B -> C in order**: All cells should be clean (green checkmark)

2. **Change x to 20 in Cell A and re-run just Cell A**: Cell B becomes stale (needs re-run)

3. **Try running C before B** (after A): You'll get a NameError because `y` doesn't exist yet - order matters!

4. **Try running B after A, then running a new cell that writes to x**: You'll see a "backward violation" - can't overwrite what earlier cells read
