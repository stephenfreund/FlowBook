# Forward Contamination

This notebook shows what happens when a cell reads a variable that was
**overwritten by a later cell** in the notebook.

## How to run

1. Run **Cell A** (`x = 1`)
2. Run **Cell C** (`x = 99`) — skipping Cell B
3. Run **Cell B** (`y = x + 1`)

## What to expect

- Cells A and C execute normally and stay **fresh**.
- When you run Cell B, it reads `x`, which was overwritten by Cell C (a later
  cell). FlowBook detects this as **forward contamination**: B's result depends
  on out-of-order state.
- Cell B is accepted but marked **stale** (amber highlight). The output area
  shows a warning explaining the contamination.
- `y` will be `100` (from `x = 99`), not `2` (from `x = 1`). The stale
  marker warns you that B's result would differ if cells ran in order.

## Recovery

Right-click Cell B and select **"Run with Previous Cell's state"**. This
re-executes B from Cell A's checkpoint (`x = 1`), producing `y = 2`. Cell B
becomes fresh again.

```python
# Cell A
x = 1
```

```python
# Cell B
y = x + 1
print(f'y = {y}')
```

```python
# Cell C
x = 99
```
