# Edit-Triggered Staleness

FlowBook tracks when you **edit** a previously executed cell and marks it
and its dependents as stale — even before you re-run it.

## How to run

### Phase 1: Run everything in order

1. Run **Cell A** (`base = 100`)
2. Run **Cell B** (`tax = base * 0.1`)
3. Run **Cell C** (`total = base + tax`)

All cells are **fresh**. `total = 110`.

### Phase 2: Edit without re-running

4. **Edit Cell A**: change `base = 100` to `base = 200` (do NOT re-run it).
5. Wait 1–2 seconds for the edit to be detected.

## What to expect

- Cell A turns **stale** because its source code no longer matches its
  last execution.
- Cells B and C also turn **stale** because they depend on `base`, which
  is written by the now-stale Cell A.
- No output changes — the cells haven't been re-run yet. The stale markers
  are a warning that the notebook state may be out of date.

### Phase 3: Re-run to refresh

6. Re-run Cell A, then B, then C.

All cells become **fresh** again with `total = 220`.

```python
# Cell A — try editing this value after running all cells
base = 100
```

```python
# Cell B
tax = base * 0.1
print(f'tax = {tax}')
```

```python
# Cell C
total = base + tax
print(f'total = {total}')
```
