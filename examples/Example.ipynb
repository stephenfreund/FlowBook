```python
import math
import random
import time
import numpy as np
```

```python
N, d = 2000, 20
pts = np.random.random((N, d))
```

```python
def pairwise_distances(points):
    """
    Given points as a list of tuples (x1,…,xd),
    return an N×N matrix of Euclidean distances.
    """
    N = len(points)
    # initialize an N×N list of zeros
    D = [[0.0]*N for _ in range(N)]
    for i in range(N):
        for j in range(i+1, N):
            # compute distance
            s = 0.0
            for xi, xj in zip(points[i], points[j]):
                diff = xi - xj
                s += diff*diff
            dist = math.sqrt(s)
            D[i][j] = dist
            D[j][i] = dist
    return D
```

```python
_start = time.time()
dists = pairwise_distances(pts)
_end = time.time()
_diff = _end - _start
print("Time = ", _diff, "sec")
```
