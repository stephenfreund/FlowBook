```python
# ======================================================================
# Generate Random Points in Multidimensional Space
#
# - Defines the number of random points (N) and dimensions (d)
# - Generates an ndarray of random points using NumPy
# - Outputs a 2D array of shape (2000, 20) representing points in 20-dimensional space
#
# Global Outputs:
#   - N: int - The number of random points to generate, set to 2000.
#   - d: int - The number of dimensions for each point, set to 20.
#   - pts: ndarray[float64, shape=(2000, 20)] - An ndarray of shape (2000, 20) containing the generated random points.
# ======================================================================

# ======================================================================

# Generate Random Points in Multidimensional Space

#

# - Defines the total number of random points and dimensions

# - Generates an array of random points using NumPy

# - Outputs an ndarray representing points in the specified dimensional space

#

# Global Outputs:

#   - N: int - The number of random points to generate, set to 2000.

#   - d: int - The number of dimensions for each random point, set to 20.

#   - pts: ndarray[float64, shape=(2000, 20)] - An array of shape (2000, 20) containing the generated random points.

# ======================================================================

# ======================================================================

# Generate Random Points in a Multidimensional Space

#

# - Defines the number of points and dimensions.

# - Generates an array of random points in a specified dimensional space.

#

# Global Outputs:

#   - N: int - The number of random points to generate, set to 2000.

#   - d: int - The number of dimensions for each point, set to 20.

#   - pts: ndarray[float64, shape=(2000, 20)] - An array of shape (2000, 20) containing the generated random points.

# ======================================================================

N, d = 2000, 20

pts = np.random.random((N, d))
```

```python
# ======================================================================
# Compute Pairwise Euclidean Distances Matrix
#
# - Calculates an N×N matrix of Euclidean distances between points.
# - Points are specified as a list of tuples containing coordinates in N-dimensional space.
# - The function utilizes a double nested loop for efficient distance computation.
#
# Global Outputs:
#   - pairwise_distances: pairwise_distances(points) -> None - An N×N matrix where each element (i, j) represents the Euclidean distance between point i and point j.
# ======================================================================

# ======================================================================

# Compute Pairwise Euclidean Distances Matrix

#

# - Calculates an N×N matrix representing distances between a list of points.

# - Points are specified as tuples of coordinates in an N-dimensional space.

# - Uses a double nested loop to efficiently compute distances between points.

#

# Global Outputs:

#   - pairwise_distances: pairwise_distances(points) -> None - An N×N matrix where each element (i, j) represents the Euclidean distance between point i and point j.

# ======================================================================

# ======================================================================

# Compute Pairwise Euclidean Distances Matrix

#

# - Calculates an N×N matrix of distances between points.

# - Points are given as a list of tuples representing coordinates.

# - Utilizes a double nested loop to compute distances efficiently.

#

# Global Outputs:

#   - pairwise_distances: pairwise_distances(points) -> None

# ======================================================================

def pairwise_distances(points):

    """

    Given points as a list of tuples (x1,…,xd),

    return an N×N matrix of Euclidean distances.

    """

    N = len(points)

    # initialize an N×N list of zeros

    D = [[0.0]*N for _ in range(N)]

    for i in range(N):

        for j in range(i+1, N):

            # compute distance

            s = 0.0

            for xi, xj in zip(points[i], points[j]):

                diff = xi - xj

                s += diff*diff

            dist = math.sqrt(s)

            D[i][j] = dist

            D[j][i] = dist

    return D
```

```python
# ======================================================================
# Measure Execution Time for Pairwise Distances Function
#
# - Records the start time before executing the pairwise_distances function.
# - Calculates the pairwise distances for a given set of points stored in 'pts'.
# - Records the end time after the execution and calculates the elapsed time.
# - Prints the total elapsed time for the execution in seconds.
#
# Global Inputs:
#   - pairwise_distances: pairwise_distances(points) -> None
#   - pts: ndarray[float64, shape=(2000, 20)]
#
# Global Outputs:
#   - end: float - Timestamp recorded after the execution of the pairwise_distances function.
#   - start: float - Timestamp recorded before the execution of the pairwise_distances function.
# ======================================================================

# ======================================================================

# Measure execution time of the pairwise_distances function

#

# - Records the time before and after the execution of the pairwise_distances function.

# - Calculates the pairwise distances for a given set of points.

# - Prints the total elapsed time for the execution in seconds.

#

# Global Inputs:

#   - pairwise_distances: pairwise_distances(points) -> None

#   - pts: ndarray[float64, shape=(2000, 20)]

#

# Global Outputs:

#   - end: float - Timestamp recorded after the execution of the pairwise_distances function.

#   - start: float - Timestamp recorded before the execution of the pairwise_distances function.

# ======================================================================

# ======================================================================

# Measure execution time of pairwise_distances function

#

# - Records the start time before executing the pairwise_distances function.

# - Calculates pairwise distances for the given points 'pts'.

# - Records the end time after the execution.

# - Prints the total elapsed time for the execution in seconds.

#

# Global Inputs:

#   - pairwise_distances: pairwise_distances(points) -> None

#   - pts: ndarray[float64, shape=(2000, 20)]

#

# Global Outputs:

#   - _ - Placeholder for the output of the pairwise_distances function, which is not used further.

#   - end: float - Timestamp recorded after the execution of pairwise_distances.

#   - start: float - Timestamp recorded before the execution of pairwise_distances.

# ======================================================================

start = time.time()

_ = pairwise_distances(pts)

end = time.time()

print("Time = ", end - start, "sec")
```
