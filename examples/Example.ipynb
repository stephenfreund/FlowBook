````python
import math
import random
import time
import numpy as np
````

````python
N, d = 2000, 20
pts = np.random.random((N, d))
````

````python
# ======================================================================
# Compute Pairwise Euclidean Distances Matrix
#
# - Calculates pairwise Euclidean distances between given points.
# - Returns an N×N matrix where element (i, j) is the distance between point i and point j.
# - Works with multi-dimensional points provided as tuples within a list.
#
# Global Outputs:
#   - pairwise_distances: pairwise_distances(points) -> None - A function that returns an N×N matrix of pairwise Euclidean distances for a given list of multi-dimensional points.
# ======================================================================

```python
def pairwise_distances(points: np.ndarray) -> np.ndarray:
    """
    Given points as a 2D numpy array,
    return an N×N matrix of Euclidean distances.
    """
    diff = points[:, np.newaxis, :] - points[np.newaxis, :, :]
    dist_matrix = np.sqrt(np.sum(diff**2, axis=-1))
    return dist_matrix
```

start = time.time()
print(start)
a = pairwise_distances(pts)
end = time.time()
diff = end - start
print("Time = ", end - start, "sec")    
````
