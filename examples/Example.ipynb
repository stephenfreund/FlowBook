```python
import math
import random
import time
import numpy as np
```

```python
N, d = 2000, 20
pts = np.random.random((N, d))
```

```python
# ======================================================================
# Compute Pairwise Euclidean Distances Matrix
#
# - Calculates pairwise Euclidean distances between given points.
# - Returns an NГ—N matrix where element (i, j) is the distance between point i and point j.
# - Works with multi-dimensional points provided as tuples within a list.
#
# Global Outputs:
#   - pairwise_distances: pairwise_distances(points) -> None - A function that returns an NГ—N matrix of pairwise Euclidean distances for a given list of multi-dimensional points.
# ======================================================================

def pairwise_distances(points):
    """
    Given points as a list of tuples (x1,вЂ¦,xd),
    return an NГ—N matrix of Euclidean distances.
    """
    N = len(points)
    # initialize an NГ—N list of zeros
    D = [[0.0]*N for _ in range(N)]
    for i in range(N):
        for j in range(i+1, N):
            # compute distance
            s = 0.0
            for xi, xj in zip(points[i], points[j]):
                diff = xi - xj
                s += diff*diff
            dist = math.sqrt(s)
            D[i][j] = dist
            D[j][i] = dist
    return np.array(D, dtype=float)

start = time.time()
print(start)
a = pairwise_distances(pts)
end = time.time()
diff = end - start
print("Time = ", end - start, "sec")    
```

```python
import numpy as np

def pairwise_distances(points):
    """
    Given points as a list of tuples (x1,вЂ¦,xd),
    return an NГ—N matrix of Euclidean distances using NumPy operations.
    """
    X = np.array(points, dtype=float)  # shape (N, d)
    # Compute squared norms for each point
    sq_norms = np.sum(X**2, axis=1)  # shape (N,)
    # Use broadcasting and matrix multiplication to compute pairwise distances
    D_sq = sq_norms[:, None] + sq_norms[None, :] - 2 * X @ X.T
    # Numerical stability: ensure no negatives due to floating-point errors
    D_sq = np.maximum(D_sq, 0.0)
    return np.sqrt(D_sq)

start = time.time()
a = pairwise_distances(pts)
end = time.time()
diff = end - start
print("Time = ", end - start, "sec")
```
