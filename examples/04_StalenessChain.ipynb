# Staleness Propagation

This notebook shows how re-executing an early cell causes **staleness to
propagate** through dependent cells.

## How to run

### Phase 1: Build the chain

1. Run **Cell A** (`x = 1`)
2. Run **Cell B** (`y = x + 10`) — reads `x`, produces `y = 11`
3. Run **Cell C** (`z = y + 100`) — reads `y`, produces `z = 111`
4. Run **Cell D** (`w = z + 1000`) — reads `z`, produces `w = 1111`

All four cells are **fresh**.

### Phase 2: Re-execute the root

5. Change Cell A to `x = 2` and re-run it.

## What to expect

- Cell A becomes fresh with the new value (`x = 2`).
- Cell B becomes **stale** because it reads `x`, which changed.
- Cell C becomes **stale** because it reads `y`, written by stale Cell B.
- Cell D becomes **stale** because it reads `z`, written by stale Cell C.

All three downstream cells turn amber. Their outputs are still visible but
may be out of date. Re-run B, then C, then D in order to refresh the chain.

```python
# Cell A — try changing this value and re-running
x = 1
```

```python
# Cell B
y = x + 10
print(f'y = {y}')
```

```python
# Cell C
z = y + 100
print(f'z = {z}')
```

```python
# Cell D
w = z + 1000
print(f'w = {w}')
```
