# Structural Tracking: Nested DataFrames

Structural tracking works with DataFrames nested inside dictionaries,
lists, and objects. The path to the DataFrame is tracked so violations
can be reported accurately.

Examples:
- `data['train']` - DataFrame in a dict
- `datasets[0]` - DataFrame in a list
- `container.df` - DataFrame as object attribute

```python
import pandas as pd

%structural_tracking warn
```

## DataFrames in Dictionaries

Common pattern: train/test splits stored in a dict.

```python
# Create train/test split
data = {
    'train': pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]}),
    'test': pd.DataFrame({'x': [7, 8], 'y': [9, 10]})
}
```

```python
# Read structure of training data
print(f"Train columns: {list(data['train'].columns)}")
print(f"Train shape: {data['train'].shape}")
```

```python
# Read structure of test data
print(f"Test columns: {list(data['test'].columns)}")
print(f"Test shape: {data['test'].shape}")
```

## DataFrames in Lists

```python
# List of DataFrames
datasets = [
    pd.DataFrame({'a': [1, 2]}),
    pd.DataFrame({'b': [3, 4]}),
    pd.DataFrame({'c': [5, 6]})
]
```

```python
# Read structure of each DataFrame
for i, df in enumerate(datasets):
    print(f"datasets[{i}] columns: {list(df.columns)}")
```

## DataFrames as Object Attributes

```python
# Simple container class
class DataContainer:
    def __init__(self):
        self.features = pd.DataFrame({'f1': [1, 2], 'f2': [3, 4]})
        self.labels = pd.Series([0, 1], name='label')

container = DataContainer()
```

```python
# Read structure of container's DataFrames
print(f"Features columns: {list(container.features.columns)}")
print(f"Labels length: {len(container.labels)}")
```

## Path Tracking

When a structural violation is detected, the full path is reported:
- `data['train']` instead of just `DataFrame`
- `datasets[0]` with the list index
- `container.features` with the attribute name

This makes it easy to identify which DataFrame caused the issue.
