```python
# Cell 1: Imports
import numpy as np
import pandas as pd
from collections import defaultdict
```

```python
# Cell 2: Create multiple numpy arrays with different dtypes
arr_float64 = np.random.randn(100000)  # ~800KB
arr_float32 = np.random.randn(100000).astype(np.float32)  # ~400KB
arr_int32 = np.random.randint(0, 1000, 100000, dtype=np.int32)  # ~400KB
arr_bool = np.random.choice([True, False], 100000)  # ~100KB
print(f'float64: {arr_float64.nbytes/1024:.1f}KB, float32: {arr_float32.nbytes/1024:.1f}KB')
print(f'int32: {arr_int32.nbytes/1024:.1f}KB, bool: {arr_bool.nbytes/1024:.1f}KB')
```

```python
# Cell 3: Create DataFrames with different column types
df_numeric = pd.DataFrame({
    'a': np.random.randn(50000),
    'b': np.random.randint(0, 100, 50000),
    'c': np.random.choice([1.0, 2.0, 3.0], 50000)
})
df_mixed = pd.DataFrame({
    'id': range(10000),
    'name': [f'item_{i}' for i in range(10000)],
    'value': np.random.randn(10000),
    'category': pd.Categorical(['A', 'B', 'C'] * 3333 + ['A'])
})
print(f'df_numeric: {df_numeric.memory_usage(deep=True).sum()/1024:.1f}KB')
print(f'df_mixed: {df_mixed.memory_usage(deep=True).sum()/1024:.1f}KB')
```

```python
# Cell 4: Create Python containers
large_list = list(range(100000))
nested_dict = {f'key_{i}': {'inner': list(range(100)), 'value': i} for i in range(1000)}
large_set = set(range(50000))
print(f'list: {len(large_list)} items, dict: {len(nested_dict)} keys, set: {len(large_set)} items')
```

```python
# Cell 5: Object array with mixed Python objects
object_array = np.array([
    {'type': 'dict', 'data': list(range(50))},
    [1, 2, 3, 'string', 4.5],
    'a string element',
    (1, 2, 3, 4, 5),
    set(range(20))
] * 1000, dtype=object)
print(f'object_array: {len(object_array)} elements')
```

```python
# Cell 6: DataFrame operations - groupby and aggregation
df_grouped = df_numeric.groupby(pd.cut(df_numeric['b'], bins=10)).agg({
    'a': ['mean', 'std', 'count'],
    'c': ['sum', 'min', 'max']
})
df_pivot = df_mixed.pivot_table(values='value', index='category', aggfunc=['mean', 'std', 'count'])
print(f'Grouped shape: {df_grouped.shape}, Pivot shape: {df_pivot.shape}')
```

```python
# Cell 7: Create views and slices (shared memory)
arr_view = arr_float64[::2]  # View, shares memory
arr_slice = arr_float64[:50000].copy()  # Copy, new memory
df_view = df_numeric[['a', 'b']]  # View in pandas
df_copy = df_numeric[['a', 'b']].copy()  # Explicit copy
print(f'view shares memory: {np.shares_memory(arr_float64, arr_view)}')
print(f'slice shares memory: {np.shares_memory(arr_float64, arr_slice)}')
```

```python
# Cell 8: Functions with closures
captured_data = np.random.randn(10000)
def compute_stats():
    return captured_data.mean(), captured_data.std()

def make_multiplier(factor):
    large_array = np.ones(5000) * factor
    def multiply(x):
        return x * large_array
    return multiply

mult_2 = make_multiplier(2)
mult_3 = make_multiplier(3)
print(f'compute_stats: {compute_stats()}')
print(f'mult_2(10): {mult_2(10).sum()}')
```

```python
# Cell 9: Large 2D array and matrix operations
matrix_a = np.random.randn(500, 500)  # ~2MB
matrix_b = np.random.randn(500, 500)  # ~2MB
matrix_product = matrix_a @ matrix_b  # ~2MB result
print(f'matrix_a: {matrix_a.nbytes/1024/1024:.1f}MB')
print(f'matrix_product: {matrix_product.nbytes/1024/1024:.1f}MB')
```

```python
# Cell 10: Series with different dtypes
series_float = pd.Series(np.random.randn(50000))
series_string = pd.Series([f'value_{i % 100}' for i in range(10000)])
series_datetime = pd.Series(pd.date_range('2020-01-01', periods=10000, freq='h'))
series_nullable = pd.Series([1, 2, None, 4, 5] * 10000, dtype='Int64')
print(f'float series: {series_float.memory_usage(deep=True)/1024:.1f}KB')
print(f'string series: {series_string.memory_usage(deep=True)/1024:.1f}KB')
print(f'datetime series: {series_datetime.memory_usage(deep=True)/1024:.1f}KB')
```

```python
# Cell 11: Complex nested structure
complex_structure = {
    'arrays': [np.random.randn(1000) for _ in range(10)],
    'dataframes': [pd.DataFrame({'x': range(500), 'y': np.random.randn(500)}) for _ in range(5)],
    'nested': {
        'level1': {
            'level2': {
                'data': list(range(1000)),
                'more_data': set(range(500))
            }
        }
    },
    'mixed_list': [1, 'two', 3.0, [4, 5], {'six': 6}] * 100
}
print(f'complex_structure has {len(complex_structure)} top-level keys')
```

```python
# Cell 12: Final summary
print('=== Memory Benchmark Complete ===')
print(f'Total arrays: 8 numpy arrays')
print(f'Total DataFrames: 4 DataFrames')
print(f'Total Series: 4 Series')
print(f'Total containers: lists, dicts, sets, nested structures')
print(f'Functions with closures: 3')
```
