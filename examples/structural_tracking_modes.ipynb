# Structural Tracking Modes

This notebook demonstrates the three structural tracking modes:
- **OFF**: No structural tracking
- **WARN**: Track and warn (default)
- **ENFORCE**: Track and block violations

```python
import pandas as pd
```

## OFF Mode

In OFF mode, structural access is not tracked at all.

```python
%structural_tracking off
```

```python
df = pd.DataFrame({'a': [1, 2, 3]})
print(f"Columns: {list(df.columns)}")
# In OFF mode, this access is not tracked
```

## WARN Mode (Default)

In WARN mode, structural access is tracked and warnings are issued
when changes would affect earlier reads, but execution continues.

```python
%structural_tracking warn
```

```python
df = pd.DataFrame({'a': [1, 2, 3]})
print(f"Columns: {list(df.columns)}")
# In WARN mode, this access IS tracked
```

## ENFORCE Mode

In ENFORCE mode, structural access is tracked and changes that
would affect earlier reads are treated as violations (blocked in monotonicity mode).

```python
%structural_tracking enforce
```

```python
df = pd.DataFrame({'a': [1, 2, 3]})
print(f"Columns: {list(df.columns)}")
# In ENFORCE mode, adding columns would be blocked
```

```python
# Check current mode
%structural_tracking
```

```python
# Reset to default
%structural_tracking warn
```
